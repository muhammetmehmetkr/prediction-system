```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("last_data.csv")
```

```python
data.head()
```

```python
X = data.drop(['region','charges'], axis = 1)
```

```python
y = data['charges']
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
model = RandomForestRegressor(n_estimators=700,max_depth=4).fit(X_train, y_train)
```

```python
model_confidence = model.score(X_test, y_test)
```

```python
print(model_confidence)
```

```python
y_pred = model.predict(X_test)
```

```python
X_ax = range(len(y_test))
```

```python
plt.figure(figsize=(15,5))
plt.plot(X_ax, y_test, label = "original")
plt.plot(X_ax, y_pred, label = "predicted")
plt.title("Test and Prediction Data")
plt.xlabel("Prediction Numbers")
plt.ylabel("Charges")
plt.legend()
plt.show()
```
